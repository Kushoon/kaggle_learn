```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
home_data = pd.read_csv("datasets/AmesHousing.csv")
pd.options.display.max_columns = None
home_data.head()
```

```python
home_data.describe()
```

```python
home_data.fillna(0, inplace=True)
```

```python
features = ['Lot Area', 'Year Built', 'Year Remod/Add', '1st Flr SF', '2nd Flr SF', 'TotRms AbvGrd']
X = home_data[features]
```

```python
y = home_data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

leaf_nodes = [5 , 10, 100, 500, 1000]

def get_mae(max_leaf_nodes, X_train, y_train, X_test, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(X_train, y_train)
    pred_values = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred_values)
    return mae

for leaf in leaf_nodes:
    my_mae = get_mae(leaf, X_train, y_train, X_test, y_test)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(leaf, my_mae))
```

```python
# Max leaf node seems like the best choice as its the lowest so lets train it on the entire dataset now
model = DecisionTreeRegressor(max_leaf_nodes = 100, random_state = 0)
model.fit(X, y)
pred_val = model.predict(X)
print(pred_val)
```
